```python
import pandas as pd
class Parent():
    def __init__(self,data = pd.read_csv(r"C:\Users\aman.kumar\Downloads\python ass data.csv")):
        self.df=data
    def DateFormatting(self):
        self.df['DATE']=pd.to_datetime(self.df['DATE'])
        self.df['DATE']=self.df['DATE'].dt.strftime('%Y-%m-%d')
        return self.df
    def AvgAccToYear(self):
        self.DateFormatting()
        new=self.df.groupby(['YEAR_OF_PASSING','FIRST_NAME','LAST_NAME','TOTAL_MARKS'],as_index=False).mean()
        new['AVG_YEARLY']=new['MARKS']
        new=new.drop(['MARKS'],axis=1)
        self.df=pd.merge(new,self.df,how='left')
        self.df.reset_index()
        return self.df
    def AvgAccToSub(self):
        self.AvgAccToYear()
        neww=self.df.groupby(['YEAR_OF_PASSING','FIRST_NAME','LAST_NAME','TOTAL_MARKS','SUBJECTS'],as_index=False).mean()
        neww['AvgSub']=neww['MARKS']
        return neww
    def grade(self,val):
        self.AvgAccToYear()
        if val>=90:
            return 'S'
        if val>=80:
            return 'A'
        if val>=70:
            return 'B'
        if val>=60:
            return 'C'
        if val>=50:
            return 'D'
        if val>=40:
            return 'E'
        if val>=30:
            return 'U'
    def gradeMapping(self):
        self.AvgAccToYear()
        self.df['GRADES']=self.df['AVG_YEARLY'].apply(self.grade)
        return self.df
val=Parent()
val.AvgAccToSub()
```
